```python
import matplotlib as mpl 
import matplotlib.pyplot as plt
import numpy as np

from sklearn.cluster import KMeans
import cv2
```

```python
image = cv2.imread('../images/panda.png')
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
X = image.reshape(-1, 3)
X = np.float32(X)
X.dtype
```

```python
X.shape
```

```python
kmeans = KMeans(n_clusters=4, n_init=20)
```

```python
kmeans.fit(X)
```

```python
segmented_img = kmeans.cluster_centers_[kmeans.labels_]
segmented_img = segmented_img.reshape(image.shape)
```

```python
plt.imshow(segmented_img / 255)
```

```python
cv2.imwrite('image1.png', cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
cv2.imwrite('image2.png', cv2.cvtColor(segmented_img.astype("uint8"), cv2.COLOR_BGR2RGB))
```
